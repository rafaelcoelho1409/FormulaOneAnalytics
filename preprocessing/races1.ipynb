```python
import pandas as pd
import os
import datetime as dt
from dateutil.relativedelta import relativedelta
pd.options.display.max_columns = 100
pd.options.display.max_rows = 100
```

```python
races = pd.read_csv("../data/races.csv")
```

```python
races.head()
```

```python
circuits = pd.read_pickle("../data/circuits.pkl")
```

```python
circuits.head()
```

```python
races_circuits = pd.merge(
    races,
    circuits,
    on = "circuitId",
    how = "left"
)
```

```python
races_circuits.head()
```

```python
races_circuits = races_circuits.rename(columns = {"name_x": "name_race", "name_y": "name_circuit"})
```

```python
races_circuits = races_circuits.rename(columns = {"url_x": "url_race", "url_y": "url_circuit"})
```

```python
races_circuits.to_pickle("../data/races_circuits.pkl")
```
